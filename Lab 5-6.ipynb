```python
days=[1,2,3,4,5,6,7]
max_t=[50,51,52,48,47,49,46]
min_t=[43,42,40,44,33,35,37]
avg_t=[45,48,48,46,40,42,41]

plt.plot(days, max_t, label="max")
plt.plot(days, min_t, label="min")
plt.plot(days, avg_t, label="average")

plt.legend(loc='best')
plt.grid()
```

```python
import matplotlib.pyplot as plt
days = ['Day 1', 'Day 2', 'Day 3', 'Day 4', 'Day 5', 'Day 6', 'Day 7', 'Day 8', 'Day 9', 'Day 10']
sales = [10, 15, 18, 22, 25, 30, 32, 35, 40, 45]
plt.plot(days, sales, marker='o', color='blue', linestyle='-', linewidth=2)
plt.title('Sales Over 10 Days')
plt.xlabel('Days')
plt.ylabel('Sales')
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt
fruits = ['Apple', 'Banana', 'Orange', 'Grapes', 'Mango']
quantity = [25, 40, 35, 50, 30]
plt.bar(fruits, quantity, color='green')
plt.title('Fruit Quantities')
plt.xlabel('Fruits')
plt.ylabel('Quantity')
plt.show()
```

```python
import matplotlib.pyplot as plt
food_types = ['Pizza', 'Pasta', 'Burger', 'Salad', 'Ice Cream']
percentage = [35, 30, 20, 10, 5]
plt.pie(percentage, labels=food_types, autopct='%1.1f%%', startangle=90, colors=['red', 'yellow', 'blue', 'green', 'purple'])
plt.title('Food Preferences')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
ages = [22, 25, 23, 27, 30, 35, 40, 28, 24, 30, 33, 31, 29, 35, 40]
plt.hist(ages, bins=5, color='orange', edgecolor='black')
plt.title('Age Distribution')
plt.xlabel('Age Groups')
plt.ylabel('Frequency')
```

```python
import matplotlib.pyplot as plt
height = [150, 160, 170, 180, 155, 165, 175, 185]
weight = [50, 60, 65, 75, 55, 65, 70, 80]
plt.scatter(height, weight, color='blue')
plt.title('Height vs Weight')
plt.xlabel('Height (cm)')
plt.ylabel('Weight (kg)')
plt.show()
```

```python
import matplotlib.pyplot as plt
salaries = [35000, 45000, 55000, 60000, 75000, 80000, 90000, 100000, 120000]
plt.boxplot(salaries, vert=False, patch_artist=True, notch=True)
plt.title('Salary Distribution')
plt.xlabel('Salary')
plt.show()
```

```python
import matplotlib.pyplot as plt
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul']
temperature = [30, 32, 35, 33, 31, 29, 28]
plt.plot(months, temperature, color='red', marker='o', linestyle='-', linewidth=2)
plt.title('Monthly Temperature', fontsize=14, fontweight='bold')
plt.xlabel('Month', fontsize=12)
plt.ylabel('Temperature (°C)', fontsize=12)
plt.grid(True, which='both', linestyle='--', linewidth=0.5)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
x = np.linspace(0, 10, 100)
y = np.sin(x)
fig, axs = plt.subplots(2, 2, figsize=(10, 8))
axs[0, 0].plot(x, y, color='blue')
axs[0, 0].set_title('Line Plot')
axs[0, 1].bar(['A', 'B', 'C'], [10, 20, 15], color='green')
axs[0, 1].set_title('Bar Plot')
axs[1, 0].hist(np.random.randn(1000), bins=30, color='orange')
axs[1, 0].set_title('Histogram')
axs[1, 1].scatter(np.random.rand(50), np.random.rand(50), color='red')
axs[1, 1].set_title('Scatter Plot')
plt.tight_layout()
plt.show()
```
