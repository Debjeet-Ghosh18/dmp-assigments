```python
import pandas as pd
data = pd.read_csv("https://raw.githubusercontent.com/plotly/datasets/master/diabetes.csv")
print(data.head())
```

```python
print("Mean:\n", data.mean())
print("\nMedian:\n", data.median())
print("\nMode:\n", data.mode().iloc[0])
```

```python
print("Standard Deviation:\n", data.std())
print("\nVariance:\n", data.var())
```

```python
print("Correlation:\n", data.corr())
print("\nCovariance:\n", data.cov())
```

```python
print("Minimum Values:\n", data.min())
print("\nMaximum Values:\n", data.max())
print("\nRange:\n", data.max() - data.min())
```

```python
print("Missing Values:\n", data.isnull().sum())
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(8, 5))
plt.hist(data['Age'], bins=20, color='skyblue', edgecolor='black')
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(8, 5))
plt.scatter(data['Glucose'], data['Insulin'], alpha=0.6, c='purple')
plt.title('Glucose vs Insulin')
plt.xlabel('Glucose')
plt.ylabel('Insulin')
plt.show()
```

```python
plt.figure(figsize=(8, 5))
plt.boxplot(data['BloodPressure'], vert=False, patch_artist=True, boxprops=dict(facecolor='lightblue'))
plt.title('Blood Pressure Spread')
plt.xlabel('Blood Pressure')
plt.show()
```

```python
import seaborn as sns 
plt.figure(figsize=(10, 8))
correlation_matrix = data.corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Heatmap')
plt.show()
```
